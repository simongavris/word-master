```python

import requests

url = 'https://scrabble.de.wordsdb.ws/letters/-/wordlength/5/type/contain/offset/0/limit/20000'
resp = requests.get(url=url)
data = resp.json()

words = "const words = {\n"
answers = "const answers = [\n"

for w in data['data']['5']:
    word = w['w'].lower()
    skip_word = False;
    for l in ['', '' ,'']:
        if l in word:
            skip_word = True
            continue
    if skip_word is True:
        continue
    words+="  "
    words+=word
    words+=": true,"
    words+="\n"
    answers+="  '"
    answers+=word
    answers+="',\n"

words += "}\n\nexport default words"
answers += "]\n\nexport default answers"

print(words)

with open("src/data/words.js",'r+') as word_file:
    data = word_file.read()
    word_file.seek(0)
    word_file.write(words)
    word_file.truncate()
    
    

with open("src/data/answers.js",'r+') as ans_file:
    data = ans_file.read()
    ans_file.seek(0)
    ans_file.write(answers)
    ans_file.truncate()
```
